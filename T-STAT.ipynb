```python
##Q1.
Estimation statistics is a branch of statistics that deals with the process of making educated guesses or inferences about the population parameters based on sample data. It involves using various techniques to estimate unknown values of population parameters, such as mean, variance, proportion, etc.

There are two main types of estimates in estimation statistics: point estimates and interval estimates.

Point Estimate: A point estimate is a single numerical value that is used to estimate the value of an unknown parameter. For example, if we want to estimate the population mean, we can use the sample mean as a point estimate. The sample mean is a single numerical value that represents the central tendency of the sample data.

Interval Estimate: An interval estimate is a range of values that is used to estimate the value of an unknown parameter. For example, if we want to estimate the population mean, we can use the sample mean and the standard error to construct a confidence interval. The confidence interval is a range of values that we are fairly confident the population mean falls within.

In general, interval estimates are preferred over point estimates because they provide more information about the uncertainty associated with the estimate. However, point estimates are often used as a quick and simple way to estimate unknown parameters, especially in situations where interval estimates are difficult or time-consuming to compute.
```

```python
##Q2.
def estimate_pop_mean(sample_mean, sample_std, sample_size):
    """
    Estimate the population mean using a sample mean and standard deviation.
    
    Args:
    sample_mean (float): The sample mean.
    sample_std (float): The sample standard deviation.
    sample_size (int): The sample size.
    
    Returns:
    float: The estimated population mean.
    """
    import math
    std_error = sample_std / math.sqrt(sample_size)
    z_score = 1.96 # for a 95% confidence interval
    margin_error = z_score * std_error
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    estimated_pop_mean = (lower_bound + upper_bound) / 2
    return estimated_pop_mean
```

```python
sample_mean = 10.2
sample_std = 1.5
sample_size = 100
estimated_pop_mean = estimate_pop_mean(sample_mean, sample_std, sample_size)
print(f"The estimated population mean is: {estimated_pop_mean}")
```

```python
##Q5.
import numpy as np
from scipy.stats import t

# Sample data from two populations
sample1 = np.array([1, 2, 3, 4, 5])
sample2 = np.array([6, 7, 8, 9, 10])

# Sample statistics
n1 = len(sample1)
n2 = len(sample2)
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the t-test statistic
se_diff = np.sqrt((std1**2 / n1) + (std2**2 / n2))
t_stat = (mean1 - mean2) / se_diff

# Calculate the degrees of freedom
df = n1 + n2 - 2

# Calculate the p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# Define the significance level
alpha = 0.05

# Conduct the hypothesis test
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
In this example, we have two sample arrays (sample1 and sample2) representing data from two different populations. We then calculate the sample statistics, including the sample sizes (n1 and n2), sample means (mean1 and mean2), and sample standard deviations (std1 and std2).

Next, we calculate the t-test statistic using the formula (mean1 - mean2) / se_diff, where se_diff is the standard error of the difference between the two sample means. We then calculate the degrees of freedom (df) using n1 + n2 - 2.

Finally, we calculate the p-value using the cumulative distribution function (CDF) of the t-distribution, and compare it to the significance level (alpha). If the p-value is less than the significance level, we reject the null hypothesis that there is no difference between the two population means, otherwise we fail to reject the null hypothesis.

Note that this is just an example script, and the actual methods used for hypothesis testing may vary depending on the specific problem and data.
```

```python
##Q7.
The general steps involved in hypothesis testing are as follows:

State the null and alternative hypotheses: The first step is to clearly define the null hypothesis (H0) and alternative hypothesis (Ha) in terms of the research question being investigated.

Choose a significance level: The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. A common choice of α is 0.05, which corresponds to a 5% chance of rejecting the null hypothesis when it is true.

Select a test statistic: The test statistic is a measure that is calculated from the sample data and is used to determine whether to reject or fail to reject the null hypothesis. The choice of test statistic depends on the type of hypothesis test being conducted and the data being analyzed.

Calculate the p-value: The p-value is the probability of obtaining a test statistic as extreme or more extreme than the one observed, assuming the null hypothesis is true. If the p-value is less than the significance level (α), then the null hypothesis is rejected.

Make a decision and draw conclusions: Based on the calculated p-value and the chosen significance level, a decision is made whether to reject or fail to reject the null hypothesis. If the null hypothesis is rejected, it can be concluded that there is evidence in favor of the alternative hypothesis. If the null hypothesis is not rejected, it can be concluded that there is not enough evidence to support the alternative hypothesis.

Interpret the results: Finally, the results are interpreted in the context of the research question being investigated. This involves discussing the implications of the findings and any limitations or assumptions of the hypothesis test.

These steps provide a general framework for conducting hypothesis tests, but the specific details and procedures may vary depending on the type of hypothesis test being conducted and the data being analyzed.
```

```python
##Q10.
import numpy as np
from scipy.stats import t

def independent_ttest(sample1, sample2, alpha=0.05):
    """Calculate the two-sample t-test for independent samples"""
    
    # Calculate means and standard deviations of samples
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)
    
    # Calculate the pooled standard deviation
    sp = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))
    
    # Calculate the t-statistic
    t_stat = (mean1 - mean2) / (sp * np.sqrt(1/n1 + 1/n2))
    
    # Calculate the degrees of freedom
    df = n1 + n2 - 2
    
    # Calculate the critical value
    cv = t.ppf(1 - alpha/2, df)
    
    # Calculate the p-value
    p = (1 - t.cdf(abs(t_stat), df)) * 2
    
    # Determine if the null hypothesis can be rejected
    if abs(t_stat) > cv:
        result = "Reject the null hypothesis"
    else:
        result = "Fail to reject the null hypothesis"
    
    # Return results
    return t_stat, df, cv, p, result
```

```python
sample1 = np.random.normal(5, 2, 20)
sample2 = np.random.normal(6, 2, 20)

t_stat, df, cv, p, result = independent_ttest(sample1, sample2)

print("t-statistic:", t_stat)
print("Degrees of freedom:", df)
print("Critical value:", cv)
print("p-value:", p)
print(result)
```

```python
##Q12.
t-statistic, also known as the t-value or test statistic, is a numerical value used in hypothesis testing to determine whether the null hypothesis should be rejected or not. It is a measure of the difference between the sample mean and the hypothesized population mean, scaled by the standard error of the sample mean.

The formula for t-statistic is as follows:

t = (x̄ - μ) / (s / √n)

where:

x̄ is the sample mean
μ is the hypothesized population mean
s is the sample standard deviation
n is the sample size
The t-statistic measures how many standard errors the sample mean is away from the hypothesized population mean. If the t-statistic is large, it suggests that the difference between the sample mean and hypothesized population mean is significant and the null hypothesis can be rejected. The significance level and degrees of freedom are used to determine the critical value of the t-statistic for hypothesis testing.



```
